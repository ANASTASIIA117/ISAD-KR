```python
# Outliers
Торг_площа = []
for Торг_площа_м2 in df['Торг_площа_м2']:
    if Торг_площа_м2 > (df['Торг_площа_м2'].mean() + 3*df['Торг_площа_м2'].std()):
        Торг_площа_м2 = (dfm['Торг_площа_м2'].mean() + 3*df['Торг_площа_м2'].std())
    elif Торг_площа_м2 < (df['Торг_площа_м2'].mean() - 3*df['Торг_площа_м2'].std()):
        Торг_площа_м2 = (dfm['Торг_площа_м2'].mean() - 3*df['Торг_площа_м2'].std())
    Торг_площа.append(Торг_площа_м2)
df['Торг_площа_м2'] = Торг_площа

Кол_во_сотрудников_ = []
for Кол_во_сотрудников in df['Кол_во_сотрудников']:
    if Кол_во_сотрудников > (df['Кол_во_сотрудников'].mean() + 3*df['Кол_во_сотрудников'].std()):
        Кол_во_сотрудников = (dfm['Кол_во_сотрудников'].mean() + 3*df['Кол_во_сотрудников'].std())
    elif Кол_во_сотрудников < (df['Кол_во_сотрудников'].mean() - 3*df['Кол_во_сотрудников'].std()):
        Кол_во_сотрудников = (dfm['Кол_во_сотрудников'].mean() - 3*df['Кол_во_сотрудников'].std())
    Кол_во_сотрудников_.append(Кол_во_сотрудников)
df['Кол_во_сотрудников'] = Кол_во_сотрудников_

Рейтинг_гугл_ = []
for Рейтинг_гугл in df['Рейтинг_гугл']:
    if Рейтинг_гугл > (df['Рейтинг_гугл'].mean() + 3*df['Рейтинг_гугл'].std()):
        Рейтинг_гугл = (dfm['Рейтинг_гугл'].mean() + 3*df['Рейтинг_гугл'].std())
    elif Рейтинг_гугл < (df['Рейтинг_гугл'].mean() - 3*df['Рейтинг_гугл'].std()):
        Рейтинг_гугл = (dfm['Рейтинг_гугл'].mean() - 3*df['Рейтинг_гугл'].std())
    Рейтинг_гугл_.append(Рейтинг_гугл)
df['Рейтинг_гугл'] = Рейтинг_гугл_

Ср_чек_ = []
for Ср_чек in df['Ср_чек']:
    if Ср_чек > (df['Ср_чек'].mean() + 3*df['Ср_чек'].std()):
        Ср_чек = (dfm['Ср_чек'].mean() + 3*df['Ср_чек'].std())
    elif Ср_чек < (df['Ср_чек'].mean() - 3*df['Ср_чек'].std()):
        Ср_чек = (dfm['Ср_чек'].mean() - 3*df['Ср_чек'].std())
    Ср_чек_.append(Ср_чек)
df['Ср_чек'] = Ср_чек_

```

```python
# Function Encoding
def encoding_char(x):
    char_var = list(set(x.columns) - set(x._get_numeric_data().columns))
    for col_names in char_var:
        f = pd.factorize(x[col_names])
        x[col_names] = pd.factorize(x[col_names])[0]
    return(x)
```

```python
# K-Means Clustering
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Using the silhouette method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 6):
  kmeans = KMeans(n_clusters = k).fit(df)  
  preds = kmeans.fit_predict(df)
  sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 6), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
 print(str(i+2) +":"+ str(sil[i]))
```
